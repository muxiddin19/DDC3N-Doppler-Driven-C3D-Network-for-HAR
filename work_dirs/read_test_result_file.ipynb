```python
import pickle
import numpy as np
from sklearn.metrics import top_k_accuracy_score
with open('result_figure_full_last.pkl', 'rb') as f:
    data = pickle.load(f)
with open('./data/figure_full/test_full.pkl', 'rb') as f:
    answer = pickle.load(f)
answer = np.array([int(ans['label']) for ans in answer])

top_3 = top_k_accuracy_score(answer, data, k=3, labels=np.arange(53))
print('Top-3 Accuracy: ', round(top_3*100, 2), '%')
```
